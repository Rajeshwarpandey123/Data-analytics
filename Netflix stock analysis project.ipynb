```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
df2=pd.read_csv("Netflix.csv")
```

```python
df2
```

```python
df2.head()
```

```python
sns.set(rc={"figure.figsize":(10,5)})
```

```python
df2["Date"]=pd.to_datetime(df2["Date"])
df2=df2.set_index('Date')
df2.head()
```

```python
sns.lineplot(x=df2.index,y=df2["Volume"],label="Volume")
plt.title("Volume of stock vs time")
```

```python
df2.plot(y=["High",'Open',"Close"],title="Netflix stock price")
```

```python
fig,(ax1,ax2,ax3)=plt.subplots(3,figsize=(15,10))
df2.groupby(df2.index.day).mean().plot(y="Volume",ax=ax1,xlabel='Day')
df2.groupby(df2.index.month).mean().plot(y="Volume",ax=ax2,xlabel='Month')
df2.groupby(df2.index.year).mean().plot(y="Volume",ax=ax3,xlabel='Year')
```
